```python
# Step 1: Import the required libraries
import matplotlib.pyplot as plt
import numpy as np

# Step 2: Create some data
x = np.linspace(0, 10)  # Generate 100 points from 0 to 10
y = np.sin(x)  # Sine function of x

# Step 3: Create a plot
plt.scatter(x, y, label='Sine Wave')  # Plot y vs x

# Step 4: Add titles and labels
plt.title('Simple Sine Wave Plot')  # Title of the plot
plt.xlabel('X Axis')  # Label for X-axis
plt.ylabel('Y Axis')  # Label for Y-axis

# Step 6: Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data=np.linspace(0,10,100)

data_x=np.sin(data)
data_y=np.cos(data)

plt.scatter(data_x,data_y,label='Sine Wave')

plt.title('Simple Sine Wave Plot')
plt.xlabel('Sine Wave')
plt.ylabel('Cosine Wave')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.linspace(0, 20, 20)

x=2*data + 1
y=data**2

plt.plot(data, x, label="Linear")
plt.plot(data, y, label="Quadratic")

plt.title("Linear and Quadratic")
plt.xlabel("X")
plt.ylabel("Y")
plt.xticks([1, 5, 10, 15, 20])
plt.yticks([5, 10, 15, 20])
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate 100 points from 0 to 10
x = np.linspace(0, 10, 100)

# Linear function y = 2x + 1
y1 = 2 * x + 1

# Quadratic function y = x^2
y2 = x ** 2

# Plot both functions with labels
plt.plot(x, y1, label='Linear Function' , marker='.')  # Linear function
plt.plot(x, y2, label='Quadratic Function', marker='.')  # Quadratic function

# Add title and labels
plt.title('Linear vs Quadratic Functions')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.linspace(0, 10, 100)

x = np.sin(data)
y = -np.sin(data)

plt.plot(data, x, label = "Sine Wave", marker='.')
plt.plot(data, y, label = "Negative wave", marker='.')

plt.title("Sine and Neg Sine Waves")

plt.xlabel("Sine Wave")
plt.ylabel("Neg Sine Wave")

plt.legend()

plt.grid()

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    'Salary' : [20000, 26000,21000],
    'Age' : [22, 26, 28]
}

employee = pd.DataFrame(data)
employee['Salary/Age'] = employee['Salary'] + employee['Age']

mean = employee['Age'].mean()

print("Salary/Age: \n",employee)
print("\nMean: ", mean)

stddev = np.std(employee['Age'])
print("\nStandard Deviation using numpy : ", stddev)

x = employee['Salary']
y = employee['Salary/Age']

plt.plot(data['Salary'],x)
plt.plot(data['Salary'],y)

plt.title('Graph of Salary and Salary/Age')
plt.xlabel('Salary')
plt.ylabel('Salary/Age')
plt.legend()
plt.show()
```
